```python
from z3 import *
import subprocess
import time

def Cexec(init_string):
    out = subprocess.check_output("./a.out %s" % init_string,shell=True,)    
    return list(map(int,out.decode('utf-8').split()))
    
    
start_time = time.time()

i1,i2,i3,i4,i6,G1,G2,G3,G4,GG1,GG2 = BitVecs('i1 i2 i3 i4 i6 G1 G2 G3 G4 GG1 GG2',32)
o1_1,o2_1,o3_1,o4_1 = BitVecs('o1_1 o2_1 o3_1 o4_1',32)
o1_2,o2_2,o3_2,o4_2 = BitVecs('o1_2 o2_2 o3_2 o4_2',32)
key1_1,key2_1,key0_1 = Bools('key1_1 key2_1 key0_1')
key1_2,key2_2,key0_2 = Bools('key1_2 key2_2 key0_2')
key4_1 = BitVec('key4_1',32)
key4_2 = BitVec('key4_2',32)



def findOutput(key1,key2,key0,key4,i1,i2,i3,i4,i6,G1,G2,G3,G4,GG1,GG2):
    op1 = If(key0, GG1 + i1, GG1 * i1)
    op2 = GG2 * i2
    op3 = G1 * i2
    op4 = G2 * i1
    op5 = If(Not((Xor((G1>10), key2))), G1 * i3, BitVecVal(0,32))
    op5 = If(Not((Xor((G1>10), key2))), op5 + GG1, op5)
    op6 = If(Not((Xor((G1>10), key2))), op5 - op4, BitVecVal(0,32))
    op5 = If(Not((Xor((G1>10), key2))), op5, i3 * i4)
    op6 = If(Not((Xor((G1>10), key2))), op6, op5 - op3)
    op6 = If(op5<op4, G2 * i4, op6)
    op6 = If(op5<op4, op6 - i3, op6)
    op17 = If(op5<op4, op6 - G2, BitVecVal(0,32))
    op17 = If(op5<op4, op17, op2 - op4)
    op2 = If(op5<op4, op2, op4 - op17)
    op17 = If(op5<op4, op17, op17 - op2)
    op7 = G1 * i4
    op8 = G2 * i3
    op9 = op1 + op2
    op10 = If(key1, op3 + op4, op3 - op4)
    op11 = op4 + op6
    op12 = op7 + op8
    op13 = op11 + G1
    o1 = op13 + key4
    op14 = i6 + op12
    o2 = op14
    op15 = G1 * op14
    op16 = op13 * G2
    op17 = If(op13==op14, op17 * op11, op17)
    op14 = If(op13==op14, op14 - op17, op14)
    op15 = If(op13==op14, op15 + op17, op15)
    op14 = If(op13==op14, op14, op13 * G1)
    op18 = op14 * G2
    op19 = op15 * op16
    op20 = op17 + op18
    op21 = G1 * op20
    op22 = op19 * G2
    op23 = op19 * G1
    op24 = op20 * G2
    op25 = op21 + op22
    op26 = op23 + op24
    op27 = op9 + op25
    o3 = op27
    op28 = op10 + op26
    o4 = op28

    o = (o1,o2,o3,o4)

    return o




j = 0

s = Tactic('smt').solver()

s.add(findOutput(key1_1,key2_1,key0_1,key4_1,i1,i2,i3,i4,i6,G1,G2,G3,G4,GG1,GG2) == out1)
s.add(findOutput(key1_2,key2_2,key0_2,key4_2,i1,i2,i3,i4,i6,G1,G2,G3,G4,GG1,GG2) == out2)


pos_set = set()
print("loop1 enter")

while s.check(out1 != out2) == sat:
    m = s.model()
    print("The key is:")
    print(str(m[key1_1]) + " " + str(m[key2_1]) + " " + str(m[key0_1]) + " " + str(m[key4_1]))
    print(str(m[key1_2]) + " " + str(m[key2_2]) + " " + str(m[key0_2]) + " " + str(m[key4_2]))
    ia = str(m[i1]) + " " + str(m[i2]) + " " + str(m[i3]) + " " + str(m[i4]) + " " + str(m[i6]) + " " + str(m[G1]) + " " + str(m[G2]) + " " + str(m[G3]) + " " + str(m[G4]) + " " + str(m[GG1]) + " " + str(m[GG2])
    [oa1,oa2,oa3,oa4] = Cexec(ia)
    s.add(findOutput(key1_1,key2_1,key0_1,key4_1,m[i1],m[i2],m[i3],m[i4],m[i6],m[G1],m[G2],m[G3],m[G4],m[GG1],m[GG2]) == oa)
    s.add(findOutput(key1_2,key2_2,key0_2,key4_2,m[i1],m[i2],m[i3],m[i4],m[i6],m[G1],m[G2],m[G3],m[G4],m[GG1],m[GG2]) == oa)
    j = j + 1

p = 0

while s.check(key1_1 == key1_2,key2_1 == key2_2,key0_1 == key0_2,key4_1 == key4_2) != unsat:   # this loop is rejecting the keys
    m = s.model()
    pos_set.add(m)
    s.add(Or(key1_1 != m[key1_1],key2_1 != m[key2_1],key0_1 != m[key0_1],key4_1 != m[key4_1]))
    s.add(Or(key1_2 != m[key1_2],key2_2 != m[key2_2],key0_2 != m[key0_2],key4_2 != m[key4_2]))
    p = p + 1

print()
print("loop2 enter")
print("Possible keys:",p)
for m in pos_set:
    print(str(m[key1_1]) + " " + str(m[key2_1]) + " " + str(m[key0_1]) + " " + str(m[key4_1]))   # this set is of possible keys
print()

pos_l = list(pos_set)

g = Tactic('smt').solver()

g.add(findOutput(key1_1,key2_1,key0_1,key4_1,i1,i2,i3,i4,i6,G1,G2,G3,G4,GG1,GG2) == out1)
g.add(findOutput(key1_2,key2_2,key0_2,key4_2,i1,i2,i3,i4,i6,G1,G2,G3,G4))

print("loop3 enter")

while len(pos_l) > 1:
    m1 = pos_l[0]
    m2 = pos_l[1]
    g.push()
    g.add(key1_1 == m1[key1_2],key2_1 == m1[key2_2],key0_1 == m1[key0_2],key4_1 == m1[key4_2])
    g.add(key1_2 == m2[key1_2],key2_2 == m2[key2_2],key0_2 == m2[key0_2],key4_2 == m2[key4_2])
    if g.check(out1 != out2) == sat:
        m = g.model()
        ia = str(m[i1]) + " " + str(m[i2]) + " " + str(m[i3]) + " " + str(m[i4]) + " " + str(m[i6]) + " " + str(m[G1]) + " " + str(m[G2]) + " " + str(m[G3]) + " " + str(m[G4]) + " " + str(m[GG1]) + " " + str(m[GG2])
        [oa1,oa2,oa3,oa4] = Cexec(ia)
        oa = tuple.tuple(BitVecVal(oa1,32),BitVecVal(oa2,32),BitVecVal(oa3,32),BitVecVal(oa4,32))
        if g.check(findOutput(m1[key1_2],m1[key2_2],m1[key0_2],m1[key4_2],m[i1],m[i2],m[i3],m[i4],m[i6],m[G1],m[G2],m[G3],m[G4],m[GG1],m[GG2]) == oa) == unsat:
            pos_l.remove(m1)
        if g.check(findOutput(m2[key1_2],m2[key2_2],m2[key0_2],m2[key4_2],m[i1],m[i2],m[i3],m[i4],m[i6],m[G1],m[G2],m[G3],m[G4],m[GG1],m[GG2]) == oa) == unsat:
            pos_l.remove(m2)
    else:
        pos_l.remove(m1)
    g.pop()

print("The final key is:")
m = pos_l[0]
print("key1 = " + str(m[key1_1]) + "\n" + "key2 = " + str(m[key2_1]) + "\n" + "key0 = " + str(m[key0_1]) + "\n" + "key4 = " + str(m[key4_1]))
print()


end_time = time.time()
taken = end_time - start_time

print("Computation took %d iterations and %f seconds." % (j, taken))  
#print("Number of times Time-Out increased: %d." % (TO_increase_count))
#print("Number of times remaining key set computed: %d." % (Key_Set_Fetch_Count))
```
